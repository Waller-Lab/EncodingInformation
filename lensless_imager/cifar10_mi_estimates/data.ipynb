```python
import numpy as np 
from cleanplots import * 
```

```python
photon_counts = [20, 40, 60, 80, 100, 150, 160, 200, 250, 300, 320]
```

Gaussian results have barely any outliers

```python
gaussian_one = []
gaussian_two = []
gaussian_three = []
gaussian_four = []
gaussian_five = []

for photon_count in photon_counts: 
    result = np.load('gaussian_mi_estimate_{}_photon_count_one_psf.npy'.format(photon_count)) 
    print(result, np.ptp(result), 'gaussian one')
    gaussian_one.append(result)
    result = np.load('gaussian_mi_estimate_{}_photon_count_two_psf.npy'.format(photon_count))
    print(result, np.ptp(result), 'gaussian two')
    gaussian_two.append(result)
    result = np.load('gaussian_mi_estimate_{}_photon_count_three_psf.npy'.format(photon_count))
    print(result, np.ptp(result), 'gaussian three')
    gaussian_three.append(result)
    result = np.load('gaussian_mi_estimate_{}_photon_count_four_psf.npy'.format(photon_count))
    print(result, np.ptp(result), 'gaussian four')
    gaussian_four.append(result)
    result = np.load('gaussian_mi_estimate_{}_photon_count_five_psf.npy'.format(photon_count))
    print(result, np.ptp(result), 'gaussian five')
    gaussian_five.append(result)
```

Pixelcnn has at most 0.17 variation at a low photon count, but overall is super consistent with its values, which is good.

```python
pixelcnn_one = []
pixelcnn_two = []
pixelcnn_three = []
pixelcnn_four = []
pixelcnn_five = []
pixelcnn_diffuser = []
pixelcnn_aperture = []

for photon_count in photon_counts:  
    result = np.load('pixelcnn_mi_estimate_{}_photon_count_one_psf.npy'.format(photon_count))
    print(result, np.ptp(result), 'pixelcnn one')
    pixelcnn_one.append(result)
    result = np.load('pixelcnn_mi_estimate_{}_photon_count_two_psf.npy'.format(photon_count))
    print(result, np.ptp(result), 'pixelcnn two')
    pixelcnn_two.append(result)
    result = np.load('pixelcnn_mi_estimate_{}_photon_count_three_psf.npy'.format(photon_count))
    print(result, np.ptp(result), 'pixelcnn three')
    pixelcnn_three.append(result)
    result = np.load('pixelcnn_mi_estimate_{}_photon_count_four_psf.npy'.format(photon_count))
    print(result, np.ptp(result), 'pixelcnn four')
    pixelcnn_four.append(result)
    result = np.load('pixelcnn_mi_estimate_{}_photon_count_five_psf.npy'.format(photon_count))
    print(result, np.ptp(result), 'pixelcnn five')
    pixelcnn_five.append(result)
    result = np.load('pixelcnn_mi_estimate_{}_photon_count_diffuser_psf.npy'.format(photon_count))
    print(result, np.ptp(result), 'pixelcnn diffuser')
    pixelcnn_diffuser.append(result)
```

## Make plots of MI now

Can use the minimum values since those are the tightest upper bounds, and are already confirmed to be consistent in range

```python
plt.plot(photon_counts, np.min(gaussian_one, axis=1), label='Gaussian, 1 PSF')
plt.plot(photon_counts, np.min(gaussian_two, axis=1), label='Gaussian, 2 PSF')
plt.plot(photon_counts, np.min(gaussian_three, axis=1), label='Gaussian, 3 PSF')
plt.plot(photon_counts, np.min(gaussian_four, axis=1), label='Gaussian, 4 PSF')
plt.plot(photon_counts, np.min(gaussian_five, axis=1), label='Gaussian, 5 PSF')
plt.legend()
plt.title("Gaussian MI Estimates for CIFAR10")
plt.xlabel("Mean Photon Count")
plt.ylabel("Estimated Mutual Information")
```

```python
plt.plot(photon_counts, np.min(pixelcnn_one, axis=1), label='PixelCNN, 1 PSF')
plt.plot(photon_counts, np.min(pixelcnn_two, axis=1), label='PixelCNN, 2 PSF')
plt.plot(photon_counts, np.min(pixelcnn_three, axis=1), label='PixelCNN, 3 PSF')
plt.plot(photon_counts, np.min(pixelcnn_four, axis=1), label='PixelCNN, 4 PSF')
plt.plot(photon_counts, np.min(pixelcnn_five, axis=1), label='PixelCNN, 5 PSF')
plt.legend()
plt.title("PixelCNN MI Estimates for CIFAR10")
plt.xlabel("Mean Photon Count")
plt.ylabel("Estimated Mutual Information")
```
