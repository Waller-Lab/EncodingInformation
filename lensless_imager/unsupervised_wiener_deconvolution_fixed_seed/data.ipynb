```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
seed_value = 10

# set photon properties 
mean_photon_count_list = [20, 40, 60, 80, 100, 150, 200, 250, 300]

# set eligible psfs
psf_names = ['one', 'two', 'three', 'four', 'five', 'diffuser', 'aperture']

# MI estimator parameters 
patch_size = 32
num_patches = 10000
bs = 500
max_epochs = 50
```

## Analyze median as well as mean

```python
mse_across_psfs = []
psnr_across_psfs = []
for psf_name in psf_names:
    mse_across_photons = []
    psnr_across_photons = []
    for photon_count in mean_photon_count_list:
        mse_list, psnr_list = np.load('recon_{}_photon_count_{}_psf.npy'.format(photon_count, psf_name))
        mean_mse = np.mean(mse_list)
        mean_psnr = np.mean(psnr_list)
        mse_across_photons.append(mean_mse)
        psnr_across_photons.append(mean_psnr)
    mse_across_psfs.append(mse_across_photons)
    psnr_across_psfs.append(psnr_across_photons)
plt.figure(figsize=(15, 5))
plt.subplot(1,2,1)
for idx, psf_name in enumerate(psf_names):
    plt.plot(mean_photon_count_list, mse_across_psfs[idx], label='{}'.format(psf_name))
plt.legend()
plt.title("Unsupervised Wiener Deconvolution MSE vs. Mean Photon Count")
plt.ylabel("Mean MSE")
plt.xlabel("Mean Photon Count")
plt.subplot(1,2,2)
for idx, psf_name in enumerate(psf_names):
    plt.plot(mean_photon_count_list, psnr_across_psfs[idx], label='{}'.format(psf_name))
plt.legend()
plt.title("Unsupervised Wiener Deconvolution PSNR vs. Mean Photon Count")
plt.ylabel("Mean PSNR")
plt.xlabel("Mean Photon Count")
```

```python
mse_across_psfs = []
psnr_across_psfs = []
for psf_name in psf_names:
    mse_across_photons = []
    psnr_across_photons = []
    for photon_count in mean_photon_count_list:
        mse_list, psnr_list = np.load('recon_{}_photon_count_{}_psf.npy'.format(photon_count, psf_name))
        median_mse = np.median(mse_list)
        median_psnr = np.median(psnr_list)
        mse_across_photons.append(median_mse)
        psnr_across_photons.append(median_psnr)
    mse_across_psfs.append(mse_across_photons)
    psnr_across_psfs.append(psnr_across_photons)
plt.figure(figsize=(15, 5))
plt.subplot(1,2,1)
for idx, psf_name in enumerate(psf_names):
    plt.plot(mean_photon_count_list, mse_across_psfs[idx], label='{}'.format(psf_name))
plt.legend()
plt.title("Unsupervised Wiener Deconvolution MSE vs. Mean Photon Count")
plt.ylabel("Median MSE")
plt.xlabel("Mean Photon Count")
plt.subplot(1,2,2)
for idx, psf_name in enumerate(psf_names):
    plt.plot(mean_photon_count_list, psnr_across_psfs[idx], label='{}'.format(psf_name))
plt.legend()
plt.title("Unsupervised Wiener Deconvolution PSNR vs. Mean Photon Count")
plt.ylabel("Median PSNR")
plt.xlabel("Mean Photon Count")
```
