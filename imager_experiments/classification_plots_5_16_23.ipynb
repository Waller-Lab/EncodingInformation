```python
diffuser_noise_array = np.array(diffuser_noise_array)
phlat_noise_array = np.array(phlat_noise_array)
unconvolved_noise_array = np.array(unconvolved_noise_array)
psf_4_noise_array = np.array(psf_4_noise_array)
```

```python
def confidence_bars(data_array, noise_length, confidence_interval=0.95):
    # can also use confidence interval 0.9 or 0.99 if want slightly different bounds
    error_lo = np.percentile(data_array, 100 * (1 - confidence_interval) / 2, axis=1)
    error_hi = np.percentile(data_array, 100 * (1 - (1 - confidence_interval) / 2), axis=1)
    mean = np.mean(data_array, axis=1)
    assert len(error_lo) == len(mean) == len(error_hi) == noise_length
    return error_lo, error_hi, mean
```

```python
error_lo, error_hi, mean = confidence_bars(diffuser_noise_array, 8)
noise_axis = [0, 25, 50, 75, 100, 125, 150, 200]
plt.title('DiffuserCam MNIST Test Accuracy With Noise')
plt.plot(noise_axis, error_lo, label='lower bound')
plt.plot(noise_axis, mean, label='mean')
plt.plot(noise_axis, error_hi, label='upper bound')
plt.legend()
plt.ylabel('Test Accuracy')
plt.xlabel('Noise Level (Photons)')

error_lo, error_hi, mean = confidence_bars(psf_4_noise_array, 8)
noise_axis = [0, 25, 50, 75, 100, 125, 150, 200]
plt.figure()
plt.title('PSF_4 MNIST Test Accuracy With Noise')
plt.plot(noise_axis, error_lo, label='lower bound')
plt.plot(noise_axis, mean, label='mean')
plt.plot(noise_axis, error_hi, label='upper bound')
plt.legend()
plt.ylabel('Test Accuracy')
plt.xlabel('Noise Level (Photons)')

error_lo, error_hi, mean = confidence_bars(unconvolved_noise_array, 8)
noise_axis = [0, 25, 50, 75, 100, 125, 150, 200]
plt.figure()
plt.title('Unconvolved MNIST Test Accuracy With Noise')
plt.plot(noise_axis, error_lo, label='lower bound')
plt.plot(noise_axis, mean, label='mean')
plt.plot(noise_axis, error_hi, label='upper bound')
plt.legend()
plt.ylabel('Test Accuracy')
plt.xlabel('Noise Level (Photons)')

error_lo, error_hi, mean = confidence_bars(phlat_noise_array, 8)
noise_axis = [0, 25, 50, 75, 100, 125, 150, 200]
plt.figure()
plt.title('Phlat MNIST Test Accuracy With Noise')
plt.plot(noise_axis, error_lo, label='lower bound')
plt.plot(noise_axis, mean, label='mean')
plt.plot(noise_axis, error_hi, label='upper bound')
plt.legend()
plt.ylabel('Test Accuracy')
plt.xlabel('Noise Level (Photons)')
```

```python
noise_axis = [0, 25, 50, 75, 100, 125, 150, 200]
fig, ax = plt.subplots(1, 1, figsize=(8, 6))

error_lo, error_hi, mean = confidence_bars(phlat_noise_array, 8)
ax.plot(noise_axis, mean, '-', label='phlat')
ax.fill_between(noise_axis, error_lo, error_hi, alpha=.4)
error_lo, error_hi, mean = confidence_bars(diffuser_noise_array, 8)
ax.plot(noise_axis, mean, '-', label='diffuser')
ax.fill_between(noise_axis, error_lo, error_hi, alpha=.4)
error_lo, error_hi, mean = confidence_bars(psf_4_noise_array, 8)
ax.plot(noise_axis, mean, '-', label='psf_4')
ax.fill_between(noise_axis, error_lo, error_hi, alpha=.4)
error_lo, error_hi, mean = confidence_bars(unconvolved_noise_array, 8)
ax.plot(noise_axis, mean, '-', label='unconvolved')
ax.fill_between(noise_axis, error_lo, error_hi, alpha=.4)
plt.xlabel('Noise Level (Photons)')
plt.ylabel('Test Accuracy')
plt.title('MNIST Simple Network Classification Accuracy with Noise')
plt.legend()
```
