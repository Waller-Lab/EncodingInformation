```python
## Can we make an approx gaussian marginal with a point source?

import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import convolve2d

# Create 2D delta function
delta = np.ones((100, 100)) * 0.05
delta[50, 50] = 40

# Create gaussian filter
std = .7
x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)
xx, yy = np.meshgrid(x, y, sparse=True)
gaussian = np.exp(-(xx**2 + yy**2) / (2*std**2))
gaussian /= gaussian.sum()

# Convolve delta with gaussian
result = convolve2d(delta, gaussian, mode='same')

# Plot the results
fig, axs = plt.subplots(2, 1, figsize=(8, 8))
axs[0].imshow(delta, cmap='inferno')
axs[0].set_title('Delta Function')
im = axs[1].imshow(gaussian, cmap='inferno')
# show colorbar
cbar = plt.colorbar(im, ax=axs[1])
axs[1].set_title('Gaussian Filter')

fig, ax = plt.subplots(1, 1, figsize=(4, 4))

_ = ax.hist(result.ravel(), bins=np.linspace(result.min(), result.max(), 40), color='k')

```
